#### Frequency Map
    Words and their counts
    Frequency of each word in the document
    Removing stopwords

```python
word_frequencies = {}
for word in text:
    if word.text not in stopwords:
            if word.text not in word_frequencies.keys():
                word_frequencies[word.text] = 1
            else:
                word_frequencies[word.text] += 1
```

```python
word_frequencies
```

#### Maximum Word Frequency
    Weighted frequency
    Each word over most occurring word
    Long sentence over short sentence

```python
maximum_frequency = max(word_frequencies.values())
```

```python
maximum_frequency
```

```python
for word in word_frequencies.keys():  
        word_frequencies[word] = (word_frequencies[word]/maximum_frequency)
```

```python
word_frequencies
```

#### Sentence Score and Ranking of Words in Each Sentence
    Sentence Tokens
    Scoring every sentence based on number of words
    Word frequency table after removing stop words

```python
sentence_list = [ sentence for sentence in text.sents ]
```

```python
sentence_list[:2]
```

```python
len(sentence_list)
```

#### Lowercase conversion

```python
[w.text.lower() for t in sentence_list for w in t ]
```

```python
sentence_scores = {}  
for sent in sentence_list:  
        for word in sent:
            if word.text.lower() in word_frequencies.keys():
                if len(sent.text.split(' ')) < 30:
                    if sent not in sentence_scores.keys():
                        sentence_scores[sent] = word_frequencies[word.text.lower()]
                    else:
                        sentence_scores[sent] += word_frequencies[word.text.lower()]
```

```python
sentence_scores
```

```python
len(sentence_scores)
```

#### Finding Top N Sentence with largest score

```python
summarized_sentences = nlargest(int(len(sentence_scores)/5), sentence_scores, key=sentence_scores.get)
```

```python
summarized_sentences
```

```python
def Summary_creator(sents1):
    for i in sents1:
        for w in summarized_sentences:
            if i == w:
                print(w)
```

```python
Summary_creator(sentence_list)
```
